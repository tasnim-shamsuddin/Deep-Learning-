```python
import numpy as np
import pandas as pd

# Replace 'na' or 'NA' or similar string versions of missing values with np.nan
df = df.replace(['na', 'NA', 'Na', 'nan', 'NaN'], np.nan)

# Convert all numeric-looking columns from string to number (ignore errors for text columns)
df = df.apply(pd.to_numeric, errors='ignore')

# Now fill missing numeric columns with their median
for feature in [feature for feature in df.columns if feature not in ['class']]:
    if pd.api.types.is_numeric_dtype(df[feature]):
        df[feature] = df[feature].fillna(df[feature].median())
```

```python
x=df.drop("class",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()
classifier.add(Dense(units=10,kernel_initializer="he_uniform",activation="relu",input_dim=170))
classifier.add(Dense(units=10,kernel_initializer="he_uniform",activation="relu"))
classifier.add(Dense(units=1,kernel_initializer="glorot_uniform",activation="sigmoid"))
```

```python
classifier.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])
```

```python
classifier.summary()
```

```python
# One-hot encode the target variable
y_train_encoded = pd.get_dummies(y_train, drop_first=True)
y_test_encoded = pd.get_dummies(y_test, drop_first=True)

model_histroy=classifier.fit(x_train,y_train_encoded,batch_size=10,epochs=20,validation_split=0.25)
```

```python
plt.plot(model_histroy.history["accuracy"])
plt.plot(model_histroy.history["val_accuracy"])
plt.title("model accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")
```
