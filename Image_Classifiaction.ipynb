```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import tensorflow as tf
import seaborn as sns
```

```python
mnist =tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_valid,x_train=x_train[:5000]/255.0,x_train[5000:]/255.0
y_valid,y_train=y_train[:5000],y_train[5000:]
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(x_train[1000],annot=True,cmap="binary")
```

```python
LAYERS=[tf.keras.layers.Flatten(input_shape=[28,28],name="inputlayer"),
        tf.keras.layers.Dense(300,activation='relu',name="hiddenlayer1"),
        tf.keras.layers.Dense(100,activation='relu',name="hiddenlayer2"),
        tf.keras.layers.Dense(10,activation="softmax",name="outputlayer")]
```

```python
from tensorflow.keras import layers, models
```

```python
model=tf.keras.models.Sequential(LAYERS)
```

```python
model.summary()
```

```python
LOSS_FUNCTION="sparse_categorical_crossentropy"
OPTIMIZER="ADAM"
METRICS=["accuracy"]
```

```python
model.compile(loss=LOSS_FUNCTION,optimizer=OPTIMIZER,metrics=METRICS)
```

```python
model.layers
```

```python
EPOCHS=5
VALIDATION_SET=(x_valid,y_valid)
history=model.fit(x_train,y_train,epochs=EPOCHS,validation_data=VALIDATION_SET)
```

```python
import time
import os

from datetime import datetime

def saveModel_path(model_dir="SAVED_MODELS/"):
    os.makedirs(model_dir, exist_ok=True)
    model_name = f"Model_{datetime.now().strftime('%Y-%m-%d-%H-%M-%S')}.keras"
    return os.path.join(model_dir, model_name)
```

```python
model.save(saveModel_path())
```

```python
model.evaluate(x_test,y_test)
```
